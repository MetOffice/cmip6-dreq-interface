# Trying to understand request variables
I would have hoped that there would be a single instance of a request variable which might point at multiple CMOR variables.  Needless to say this is not true: the mapping is complicated and obscure. It might be better to think of it as records in a database, so nothing is actually an object and names occur multiple times.  But, really, I've given up trying to work it out.

```python
from collections import defaultdict
from dqi.util import load_from_dqroot as ldq

dq = ldq()
```

```python
def pmvc(dq):
    # print mip variable count: mip -> rv(mult)
    mmap = defaultdict(lambda: defaultdict(set))
    for rv in dq.coll['requestVar'].items:
        mmap[rv.mip][rv.label].add(rv.uid)
    for (mn, rvm) in mmap.iteritems():
        for (rvn, ids) in rvm.iteritems():
            if len(ids) > 1:
                print "{} -> {}({})".format(mn, rvn, len(ids))

pmvc(dq)
```

```python
def pmvmults(dq):
    # print number of multiply-defined variables by MIP
    mvc = defaultdict(lambda: defaultdict(lambda: -1))
    for rv in dq.coll['requestVar'].items:
        mvc[rv.mip][rv.label] += 1
    for (mip, vc) in mvc.iteritems():
        c = sum(1 for i in vc.values() if i > 0)
        if c > 0:
            print "{} {}".format(mip, c)

pmvmults(dq)
```

```python
def pvmc(dq):
    # print rvs which occur in more than one MIP
    vmmips = defaultdict(set)

    for rv in dq.coll['requestVar'].items:
        vmmips[rv.label].add(rv.mip)

    for (vn, vmips) in vmmips.iteritems():
        if len(vmips) > 1:
            print "{}: {}".format(vn, len(vmips))

pvmc(dq)
```
