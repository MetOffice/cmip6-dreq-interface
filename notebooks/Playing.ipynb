```python
from dreqPy.dreq import loadDreq
```

```python
dq = loadDreq()
```

```python
dq.coll.keys()
```

```python
def checkAtts(d):
    for (k, s) in d.coll.iteritems():
        print "checking {}".format(k)
        for a in s.attDefn.keys():
            for i in s.items:
                if not hasattr(i, a):
                    print "failed at {} with {}".format(k, a)
                    return False
    return True

checkAtts(dq)
```

```python
dq.coll['var'].attDefn
```

```python
def varLabels(d):
    return [v.label for v in d.coll['var'].items]

varLabels(dq)
```

```python
def check_item_inxs(d):
    inx = d.inx
    for coll in dq.coll.values():
        for item in coll.items:
            if item._inx is not d.inx:
                return False
    return True

check_item_inxs(dq)
```

```python
def rl_by_mip(d, m):
    r = set()
    for i in d.coll['objectiveLink'].items:
        if i.label == m:
            r.add(dq.inx.uid[i.rid])
    return r

rl_by_mip(dq, 'C4MIP')
```

```python
def cvls_for_mip(mip, dq):
    # CMOR variables for a given MIP name
    return [dq.inx.uid[rv.vid].label
            for rv in dq.coll['requestVar'].items
            if rv.mip == mip]

cvl_mapping = {m: cvls_for_mip(m, dq) for m in [mip.label for mip in dq.coll['mip'].items]}
```
