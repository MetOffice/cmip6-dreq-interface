## Just following the example
This won't find experiments but should find MIP variables

```python
def mipcmvids(mipname):
    # Ids of CMORvars which link to mipname
    #
    # request groups that link to the MIP
    rgs = set(dq.inx.uid[l.refid]
              for l in (dl for dl in dq.coll['requestLink'].items
                        if dl.mip == mipname))
    # and the variable names of those groups
    return set(dq.inx.uid[rv].vid
               for rvs in (dq.inx.iref_by_sect[vg.uid].a['requestVar']
                           for vg in rgs)
               for rv in rvs)

def mipvars(mipname):
    return set(dq.inx.uid[vid].label
               for vid in mipcmvids(mipname))
```

## More randomness

```python
def mipris(mipname):
    # request item ids that link to the mip
    ris = set(ri
              for rlid in (rl.uid
                           for rl in dq.coll['requestLink'].items
                           if rl.mip == mipname)
              for ri in dq.inx.iref_by_sect[rlid].a['requestItem'])
    return ris

def mipesids(mipname):
    # the ids of experimenty things which link to mipname
    return set(dq.inx.uid[ri].esid
               for ri in mipris(mipname))

{mipname: tuple((it._h.label, it.label)
                for it in (dq.inx.uid[x] for x in mipesids(mipname)))
 for mipname in dq.inx.mip.uid.keys()}
```

```python
# A function to find all the types (sections) that refer to a given type
# This should help chasing links
#
def referring_dqtypes(dqtype):
    result = set()
    for rts in (dq.inx.iref_by_sect[ob.uid].a.keys()
                for ob in dq.coll[dqtype].items):
        result.update(rts)
    return result

{dqtype: referring_dqtypes(dqtype)
 for dqtype in dq.coll.keys()}
```

## Another try
This is based on reading `scope.py`

```python
def exid2rqlids(ex, pmax=2):
    # Approximately from rqlByExpt in scope.py
    # ex is the UID of an experiment (not its name!)
    #
    exi = dq.inx.experiment.uid[ex] # the experiment object
    assert exi._h.label == 'experiment' # it must be an experiment
    exset = set((ex, exi.egid, exi.mip)) # set of ids of things to look for
    riids = mipris(exi.mip) # set of request item UIDs for the mip
    
    # xris is a set of request items which we like
    xris = set(i for i in (dq.inx.uid[ri] for ri in riids)
                 if i.preset <= pmax and i.esid in exset)
    # xrqlids is the ids of rqls we like (remove the remarks without noise)
    xrqlids = set(xri.rlid for xri in xris
                  if dq.inx.uid[xri.rlid]._h.label != 'remarks')
    # And this is the answer we are looking for
    return xrqlids
    
def rqlids2rgids(rqlids):
    # return the request group Ids for a bunch of request link ids
    # this is based on part of the mipvars function above
    return set(dq.inx.uid[l.refid]
               for l in (dq.inx.uid[rqlid]
                         for rqlid in rqlids))

def rgids2cmvids(rgids):
    # return the CMORvar ids (note not labels) for a bunch of rgids
    return set(dq.inx.uid[rv].vid
               for rvs in (dq.inx.iref_by_sect[vg.uid].a['requestVar']
                           for vg in rgids)
               for rv in rvs)

# And finally
#
def exid2cmvids(exid):
    return rgids2cmvids(rqlids2rgids(exid2rqlids(exid)))
```

```python
{exname: tuple(set(dq.inx.uid[cmvid].label
                   for cmvid in exid2cmvids(exid))
               for exid in dq.inx.experiment.label[exname])
 for exname in dq.inx.experiment.label.keys()}
```
